```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "final_psafe_dataset.csv"
df = pd.read_csv(file_path)
```

```python
df_unique = df.drop_duplicates(subset=['rid'])
```

```python
df_unique
```

```python
df_unique.to_csv("df_unique.csv")
```

```python
plt.figure(figsize=(12, 6))

# Gender
plt.subplot(231)
df_unique['gender'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=0, fontsize=10)
plt.title("Gender Distribution")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count")

# Age
plt.subplot(232)
df_unique['age'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(5), labels=["<18", "18-30", "31-40", "41-50", "50-65"], rotation=45, fontsize=10)
plt.title("Age Group Distribution")
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Count")

# Education
plt.subplot(233)
education_labels = {0: "No Education", 1: "Primary", 2: "Secondary", 3: "Higher", 4: "Master"}
education_counts = df_unique['education'].value_counts().reindex(range(5), fill_value=0)  # Ensure all categories appear
education_counts.plot(kind='bar')
plt.xticks(ticks=range(5), labels=[education_labels[i] for i in range(5)], rotation=45, fontsize=10)
plt.title("Education Level Distribution")
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Count")

# Employment
plt.subplot(234)
df_unique['employment'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(4), labels=["Retired", "Student", "Unemployed", "Employed"], rotation=45, fontsize=10)
plt.title("Employment Status Distribution")
plt.xlabel("Employment Status", fontsize=12)
plt.ylabel("Count")

# Income
plt.subplot(235)
df_unique['income'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(5), labels=["No Income", "<1500", "1500-2500", "2500-3500", ">=3500"], rotation=45, fontsize=10)
plt.title("Income Level Distribution")
plt.xlabel("Income Level (Euros/Month)", fontsize=12)
plt.ylabel("Count")

plt.tight_layout()
plt.savefig("census_data.png", dpi=600)
plt.show()
```

```python
df_unique['license_own'] = df_unique['license_own'].replace({"Yes": "1", "No": "0", "Ja": "1", "Nein": "0"})
```

```python
plt.figure(figsize=(12, 6))

# license
plt.subplot(231)
df_unique['license_own'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(3), labels=["0", "1", "Prefer not to answer"], rotation=0)
plt.title("Driving License Ownership")
plt.xlabel("Owns a License")
plt.ylabel("Count")

# Car
plt.subplot(232)
df_unique['car_own'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(4), labels=['0', '1', '2', '3'], rotation=45)
plt.title("Car Ownership")
plt.xlabel("Number of Cars in The Household")
plt.ylabel("Count")

# Motorcycle
plt.subplot(233)
df_unique['moto_own'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(3), labels=['0', '1', '2'], rotation=45)
plt.title("Motorcycle Ownership")
plt.xlabel("Number of Motorcycles in The Household")
plt.ylabel("Count")

# Bike
plt.subplot(234)
df_unique['bike_own'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(6), labels=['0', '1', '2', '3', '4', '5'], rotation=45)
plt.title("Bicycle Ownership")
plt.xlabel("Number of Bicycles in The Household")
plt.ylabel("Count")

# Escooter
plt.subplot(235)
df_unique['escooter_own'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(2), labels=['0', '1'], rotation=45)
plt.title("E-scooter Ownership")
plt.xlabel("Number of E-scooters in The Household")
plt.ylabel("Count")

plt.tight_layout()
plt.savefig("Vehicle_ownership.png", dpi=300)
plt.show()
```

```python
from matplotlib.patches import Patch
plt.figure(figsize=(9, 5))

# Car
plt.subplot(231)
df_unique['car_frequency'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(4), labels=['1', '2', '3', '4'], rotation=0)
plt.title("Car Usage")
plt.xlabel("Frequency")
plt.ylabel("Count")

# bike
plt.subplot(232)
df_unique['bike_frequency'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(4), labels=['1', '2', '3', '4'], rotation=0)
plt.title("Bicycle Usage")
plt.xlabel("Frequency")
plt.ylabel("Count")

#E-scooter
plt.subplot(233)
df_unique['escooter_frequency'].value_counts().sort_index().plot(kind='bar')
plt.xticks(ticks=range(4), labels=['1', '2', '3', '4'], rotation=0)
plt.title("E-scooter Usage")
plt.xlabel("Frequency")
plt.ylabel("Count")


# Create manual legend handles
legend_labels = ["1 = Rarely", "2 = Occasionally", "3 = Often", "4 = Frequently"]
legend_handles = [Patch(label=label) for label in legend_labels]

# Add custom legend
plt.figlegend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1, 1),
              fontsize=10)

plt.tight_layout()
plt.savefig("Vehicle_usage_frequencies.png", dpi=300)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_stress_dataset.csv')
```

```python
stressor_cols = ['obstacles', 'pavement', 'parking', 'no_crossing', 'bike_lane', 'road_marking', 'speed', 'traffic']
```

```python
label_map = {
    'obstacles': 'Obstacles on the Path',
    'pavement': 'Poor Pavement Quality',
    'parking': 'Existance of a Parking Lane',
    'no_crossing': 'Lack of Safe Crossings',
    'bike_lane': 'Absence of Bike Lanes',
    'road_marking': 'Missing Road Markings',
    'speed': 'High Speed Limit',
    'traffic': 'Heavy Traffic'
}
```

```python
for mode in df['mode'].unique():
    fig, ax = plt.subplots(figsize=(8.27, 4))  # A4 width, landscape-ish height

    subset = df[df['mode'] == mode]
    counts = subset[stressor_cols].sum()
    counts = counts.rename(label_map)  # Apply descriptive labels
    counts = counts.sort_values()

    counts.plot(kind='barh', ax=ax, color='indianred')

    ax.set_title(f'Stressor Counts - {mode.capitalize()}', fontsize=12)
    ax.set_xlabel('Number of Respondents Stressed', fontsize=10)
    ax.set_ylabel('')  # optional: remove y-axis label
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)
    ax.grid(axis='x', linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.savefig(f'Stressors_{mode}.png', dpi=600)
    plt.show()
```
